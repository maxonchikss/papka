```python
import math
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(1, 10, 10)
y = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x ** 2)) / np.log(abs(2*x))

z = y[:5]

plt.plot(x, y, c = "y")
plt.show()
plt.scatter(np.linspace(1,5,5), z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz
from scipy.integrate import *

x = np.arange(0.0, 10.1, 1)
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

plt.plot(x, y, c = "y")
plt.fill_between(x, y)
plt.show()

area = trapz(y)
print(area)
```

Задание 4.3.4

```python
import yfinance as yf
import matplotlib.pyplot as plt

data = yf.download(['MSFT', 'AAPL', 'GOOGL'], "2021-01-01", "2021-12-31")

normalized_data = data['Close'] / data['Close'].iloc[0] * 100


normalized_data.plot()
plt.grid(linewidth=0.1)
plt.xlabel('Дата')
plt.ylabel('Нормализованная цена акций (%)')
plt.title('Сравнение акций Apple, Google и Microsoft')
plt.legend(fontsize=8, labels=['Apple', 'Google', 'Microsoft'])
plt.show()
```

Задание 4.3.5

```python
import numpy as np

x = float(input("Введите первое число: "))
y = float(input("Введите второе число: "))

print("1-Сумма\n2-Вычитание\n3-Умножение\n4-Деление")
print("5-Экспонента суммы\n6-Синус суммы\n7-Косинус суммые\n8-Первое число в степени второго")

action = input()

if action == "1":
    print(x+y)
if action == "2":
    print(x-y)
if action == "3":
    print(x*y)
if action == "4":
    print(x/y)
if action == "5":
    print(np.exp(x+y))
if action == "6":
    print(np.sin(x+y))
if action == "7":
    print(np.cos(x+y))
if action == "8":
    print(np.power(x, y))
```
